- Translation in English

```python
# translate columns name
df.columns = ['make', 'model', 'energy', 'car_classification', 'cylinder_capacity', 
              'market_category', 'tax_horsepower', 'max_horsepower', 'electrical_nominal_power',
              'weight', 'weight_horsepower_ratio', 'transmission_type', 'no_gears', 'bonus_malus',
              'low_speed_fuel_consumption', 'average_speed_fuel_consumption', 'high_speed_fuel_consumption',
              'very_high_speed_fuel_consumption', 'combined_speed_fuel_consumption', 'elect_conso',
              'electrical_autonomy', 'city_electrical_autonomy', 'co2_emission']
```

```python
energy = {"ESSENCE" : "gasoline",
          "ELECTRIC" : "electric",
          "ELEC+ESSENC HR": "phev_gas",
          "ESS+ELEC HNR" : "hev_gas",
          "GAZOLE" : "diesel", 
          'GAZ+ELEC HNR' : "phev_diesel",
          "ELEC+GAZOLE HR" : "hev_diesel",
          "ESS+G.P.L." : "autogas",
          "SUPERETHANOL" : "e85",
          "GAZ NAT.VEH" : "natural_gas_vehicle"
          }
df["energy"] = df["energy"].map(energy)
```

```python
# translate classification of the car
car_classification = {"COMBISPACE" : "leisure_activity_vehicle",
                      "TS TERRAINS/CHEMINS" : "crossover_suv",
                      "MONOSPACE COMPACT" : "compact_multi_purpose_vehicle",
                      "BERLINE" : "saloon",
                      "BREAK" : "estate",
                      "COUPE" : "coupe",
                      "MINIBUS" : "minibus",
                      "CABRIOLET" : "convertible",
                      "MONOSPACE" : "multi_purpose_vehicle",
                      "MINISPACE" : "microvan"
}
df["car_classification"] = df["car_classification"].map(car_classification)
```

```python
market_category = {
    "INFERIEURE": "low",
    "MOYENNE INFERIEURE" : "average_low",
    "MOYENNE SUPERIEURE" : "average_high",
    "LUXE": "luxury",
    "SUPERIEURE" : "high",
    "ECONOMIQUE" : "economy"
    }
df["market_category"] = df["market_category"].map(market_category)
```

```python
transmission_type = {
    'AUTOMATIQUE' : "automatic",
    'VARIATION ELECTRIQUE' : 'electrical_variable_transmission', 
    'MECANIQUE' : 'manual',
    'VARIATION CONTINUE' : 'continuously_variable_transmission'
}

df["transmission_type"] = df["transmission_type"].map(transmission_type)
```

```python
bonus_malus = {"Bonus" : "bonus",
               "Malus" : "malus",
               "AUTRES" : "others",
               "Neutre": "neutral"}

df["bonus_malus"] = df["bonus_malus"].map(bonus_malus)
```

```python
# from uppercase to lowercase
strings = list(df.dtypes[df.dtypes == "object"].index)
for i in strings:
    df[i] = df[i].str.lower().str.replace(" ", "_")
```

```python
df["make"] = df["make"].replace({'rolls_royc' : "rolls_royce"})
```
